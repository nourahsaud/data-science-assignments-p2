## Rain 

```python
# Note 

# x --> future state "tomorrow"
# x-1 --> today
# x-2 --> yesterday


# Create function to perform our heuristic
preds = []

def heuristic(df):
    
    
    # If either of last two days == True then check the 70 <= TMAX <= 30 predict true
    for x in range(len(df)):
        if (df.iloc[x-1]["RAIN"] == True) and (df.iloc[x-2]["RAIN"] == True):
            if (df.iloc[x]['TMAX'] <= 65) and (df.iloc[x]['TMIN'] >= 30):
                if (df.iloc[x-1]['PRCP'] >= 0.015) or (df.iloc[x-2]['PRCP'] >= 0.1):
                    preds.append(True)
                else:
                    preds.append(False) 
            else:
                preds.append(False)
        else:
            # Predict false if the above is not true
            preds.append(False)
    return preds
```

```python
def calc_confuse(df):
    
    "Calculate all possible results of a confusion matrix"

    # Hold all possible values and set to zero
    FP = np.zeros(len(df))
    TP = np.zeros(len(df))
    FN = np.zeros(len(df))
    TN = np.zeros(len(df))
    
    for x in range(len(df)):
        
        # True Positive
        if (df["RAIN"].iloc[x] == True) & (df["preds"].iloc[x] == True):
            TP[x] = 1
        # True Negative
        elif (df["RAIN"].iloc[x] == False) & (df["preds"].iloc[x] == False):
            TN[x] = 1
        # False Negative
        elif (df["RAIN"].iloc[x] == True) & (df["preds"].iloc[x] == False):
            FN[x] = 1
        # False Positive
        else:
            FP[x] = 1
    
    return FP, TP, FN, TN
```

```python
df["preds"] = heuristic(df) # Create a column of predictions

confuse_ls = [w,x,y,z] = calc_confuse(df) # Calculate confusion matrix stats
col_names = ["FP", "TP", "FN", "TN"] # Create column name ls

# Create new columns
for x in range(len(confuse_ls)):
    df[col_names[x]] = confuse_ls[x]

# Look at 10 random rows to determine the accuracy
df.sample(10)

```

```python
rain = []
for x in range(len(df)):
    rain.append(df.iloc[x]['RAIN'])
```

```python
# Precision 
precision = (sum (df['TP']) / (sum(df['TP']) + sum(df['FP'])))

# Recall
recall = (sum (df['TP']) / (sum(df['TP']) + sum(df['FN'])))

# Calculate Accuracy
accuracy = ((sum(df['TP']) + sum(df['TN'])) / len(df))

# print 
print('precision = ', precision)
print('recall = ', recall)
print('accuracy = ', accuracy)
```

```python
# 
pre = precision_score(rain, preds)
rec = recall_score(rain, preds)
acc = accuracy_score(rain, preds)

# print 
print('precision = ', pre)
print('recall = ', rec)
print('accuracy = ', acc) 
```

```python
def mean_val(df):
    mean = []
    for x in range(len(df)):
        mean.append(0.106)
    return mean
# Apply mean_val
df["mean"] = mean_val(df)
df.head(5)
```

```python
# Create function to perform our heuristic


def heuristic(df):
    
    preds = []
    
    for x in range(len(df)):
        if (df.iloc[x]['TMIN'] >= 50) | (df.iloc[x]['TMAX'] >= 50):
            preds.append(0)
        elif (df.iloc[x]['TMIN'] >= 40) | (df.iloc[x]['TMAX'] >= 40):
            preds.append(0.02)
        else:
            # Predict false if the above is not true
            preds.append(0.01)
                
    return preds
```

```python
# Apply Heuristic
df["preds"] = heuristic(df)
```

```python
# Create a Mean Squared Error Function for HA model
def mse(y_true, y_pred):
    return np.square(np.subtract(y_true,y_pred)).mean()
ha = mse(df['PRCP'], df['preds'])
mean = mse(df['PRCP'], df['mean'])

print('HA model = ', ha )
print('mean = ', mean )
```
